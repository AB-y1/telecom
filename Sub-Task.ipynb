```python
def plot_top_values_histogram(column, top_n=10):
    # Get value counts for the column
    value_counts = column.value_counts()

    # Selecting only the top N values
    top_values = value_counts.head(top_n)

    # Plotting the histogram
    plt.bar(top_values.index, top_values.values)
    
    if top_n >= 5:
        plt.xticks(rotation=45, ha='right')

    # Adding labels and title
    plt.xlabel('Unique Values')
    plt.ylabel('Count')
    plt.title(f'Histogram of Top {top_n} Values Counts for {column.name}')

    # Display the plot
    plt.show()
```

```python
manufacturer_column = 'Handset Manufacturer'
handset_column = 'Handset Type'

# Grouping by manufacturer and counting occurrences of each handset
grouped_data = telecom.groupby(manufacturer_column)[handset_column].value_counts()

# Identifying the top 3 handset manufacturers
top_manufacturers = telecom[manufacturer_column].value_counts().head(3).index

# Initializing a dictionary to store the top 5 handsets for each manufacturer
top_handsets_per_manufacturer = {}

# Iterating over the top handset manufacturers
for manufacturer in top_manufacturers:
    # Retrieving the top 5 handsets for each manufacturer
    top_handsets = grouped_data[manufacturer].head(5)
    top_handsets_per_manufacturer[manufacturer] = top_handsets

# Printing the top 5 handsets for each manufacturer
for manufacturer, top_handsets in top_handsets_per_manufacturer.items():
    print(f"Top 5 handsets for {manufacturer}:\n{top_handsets}\n")
```

```python
top_manufacturers = telecom[manufacturer_column].value_counts().head(3).index
```

```python
filtered_df = telecom[telecom[manufacturer_column].isin(top_manufacturers)]
```

```python
grouped_data = filtered_df.groupby([manufacturer_column, handset_column]).size().reset_index(name='count')
top_5_per_manufacturer = grouped_data.groupby(manufacturer_column).apply(lambda group: group.nlargest(5, 'count')).reset_index(drop=True)
```

```python

# Useing seaborn to create a grouped bar plot
sns.catplot(x=handset_column, y='count', hue=manufacturer_column, data=top_5_per_manufacturer, kind='bar', height=6, aspect=2)

# Adding labels and title
plt.xlabel('Handset')
plt.ylabel('Count')
plt.title('Top 5 Handsets per Top 3 Handset Manufacturers')
plt.xticks(rotation=45, ha='right')


# Display the plot
plt.show()
```
