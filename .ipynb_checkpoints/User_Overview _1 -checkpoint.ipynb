```python
# delete null rows
telecom.dropna(subset=['Bearer Id'], inplace=True)
```

```python
null_count = telecom['Bearer Id'].isnull().sum()
null_count
```

```python
total_dl = telecom['Total DL (Bytes)'].sum()
total_ul = telecom[ 'Total UL (Bytes)'].sum()
print (total_dl,"and", total_ul)
```

```python
# Aggregate per user
user_aggregated = telecom.groupby('MSISDN/Number').agg({
    'Bearer Id': 'count',               # Number of xDR sessions
    'Dur. (ms)': 'sum',      # Total session duration
    'Total DL (Bytes)': 'sum',          # Total download data
    'Total UL (Bytes)': 'sum'           # Total upload data
})
user_aggregated
```

```python

# Rename columns for clarity
user_aggregated.columns = ['Number of xDR sessions', 'Total Session Duration (s)',
                           'Total Download (Bytes)', 'Total Upload (Bytes)']

user_aggregated
```

```python
subste_aggregated = user_aggregated.head(50000)
```

```python
subste_aggregated.dtypes
```

```python
# Plotting Number of xDR sessions per user
plt.figure(figsize=(10, 6))
plt.bar(subste_aggregated.index, subste_aggregated['Number of xDR sessions'], color='skyblue')
plt.xlabel('User')
plt.ylabel('Number of xDR sessions')
plt.title('Number of xDR sessions per User')
plt.show()
```

```python


# Plotting Number of xDR sessions per user
plt.figure(figsize=(10, 6))
plt.bar(subste_aggregated.index, subste_aggregated['Number of xDR sessions'], color='skyblue')
plt.xlabel('User')
plt.ylabel('Number of xDR sessions')
plt.title('Number of xDR sessions per User')
plt.show()

# Plotting Total Session Duration, Download, and Upload data per user
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 12))

# Total Session Duration
axes[0].barh(subste_aggregated.index, subste_aggregated['Total Session Duration (s)'], color='lightcoral')
axes[0].set_xlabel('Total Session Duration (s)')
axes[0].set_ylabel('User')
axes[0].set_title('Total Session Duration per User')

# Total Download Data
axes[1].barh(subste_aggregated.index, subste_aggregated['Total Download (Bytes)'], color='lightgreen')
axes[1].set_xlabel('Total Download (Bytes)')
axes[1].set_ylabel('User')
axes[1].set_title('Total Download Data per User')

# Total Upload Data
axes[2].barh(subste_aggregated.index, subste_aggregated['Total Upload (Bytes)'], color='lightskyblue')
axes[2].set_xlabel('Total Upload (Bytes)')
axes[2].set_ylabel('User')
axes[2].set_title('Total Upload Data per User')

plt.tight_layout()
plt.show()
```
